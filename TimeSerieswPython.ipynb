```python
import seaborn as sn
import statsmodels.api as sm
import pandas as pd
import numpy as np
import statsmodels.api as sm
from scipy import stats
import matplotlib.pyplot as plt
from pylab import rcParams

%matplotlib inline
```

```python
df = pd.read_json('data/logins.json', convert_dates=[0])
df = df.set_index(0)
df['count']=np.ones(len(df))
df_hourly = df.resample('1H').apply(sum)
df_hourly.dropna(inplace=True)

res = sm.tsa.seasonal_decompose(df_hourly, freq=60)
rcParams['figure.figsize'] = 12, 7
plt.plot = res.plot()
```
